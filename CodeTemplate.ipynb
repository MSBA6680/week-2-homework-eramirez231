```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker = "AAPL"

df = yf.Ticker(ticker).history(start="2013-08-01", end="2023-08-01")

df['Returns'] = df['Close'].pct_change()
df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)

plt.figure(figsize=(14,10))

plt.subplot(3,1,1)
plt.plot(df.index, df['Close'])
plt.title(f'{ticker} Price')

plt.subplot(3,1,2)
plt.plot(df.index, df['Returns'])
plt.title(f'{ticker} Daily Returns')

plt.subplot(3,1,3)
plt.plot(df.index, df['Volatility'])
plt.title(f'{ticker} Annualized Volatility')

plt.tight_layout()
plt.show()
```
